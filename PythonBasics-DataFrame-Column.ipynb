```python
#Python Basics
#We can create a new data frame consisting of one column

import pandas as Behdad
#Make sure that the arrays must all be same length
ColumData1={"Measure1":["a1", [1,2,3,4,5], "Creating Data", "The keys> "],\
           "Measure2": ["ZZ","WW", [23, 1], "2435G7"],\
           "Measure3":["DataFrame", "Behdad"*2, "b1", "c1"], "Measure4":[1232, 3432, 788,46554],\
           "Measure5": ["er","WW", 34, "rf456"],"Measure6": ["34ZZ","43WW", "tyure", "24SS87"]}
#The keys correspond to the column labels
#The values or lists corresponding to the rows
#We then cast the dictionary to a data frame using the function data frame
ColumData12=Behdad.DataFrame(ColumData1)
print(ColumData12)


#We can create a new data frame consisting of one column
#We can create a new data frame consisting of one column.
#We just put the data frame name, in this case, ColumData12, and the name of the column header enclosed in double brackets.
NewData1=ColumData12[["Measure1"]]
print(NewData1)
NewData2=ColumData12[["Measure2"]]
print(NewData2)
NewData3=ColumData12[["Measure3"]]
print(NewData3)
NewData4=ColumData12[["Measure4"]]
print(NewData4)
NewData5=ColumData12[["Measure5"]]
print(NewData5)
NewData6=ColumData12[["Measure6"]]
print(NewData6)
ColumData12.head()
```
